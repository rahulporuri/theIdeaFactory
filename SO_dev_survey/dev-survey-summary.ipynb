```python
import glob

import pandas as pd

list_of_files = glob.glob("/Users/rahulporuri/Downloads/SO-survey-data/*.csv")
```

```python
print(list_of_files)
```

```python
def get_column_data(filename, column_name):
    temp_df = pd.read_table(filename, sep=",", engine="python")
    return temp_df[column_name].astype("category")#.value_counts()
```

```python
for file in list_of_files:
    temp_df = pd.read_table(file, sep=",", engine="python")
    col_wt_country_data = temp_df.columns.str.lower().str.contains("country") | temp_df.columns.str.lower().str.contains("region")
    country_data = temp_df[temp_df.columns[col_wt_country_data].get_values()]
    print(file.split("/")[-1].split("_")[0], country_data.columns)
```

```python
df_2011 = get_column_data("/Users/rahulporuri/Downloads/SO-survey-data/2011_Stack_Overflow_Survey_Results.csv",
                           "What Country or Region do you live in?")

df_2012 = get_column_data("/Users/rahulporuri/Downloads/SO-survey-data/2012_Stack_Overflow_Survey_Results.csv",
                           "What Country or Region do you live in?")

df_2013 = get_column_data("/Users/rahulporuri/Downloads/SO-survey-data/2013 Stack Overflow Survey Responses.csv",
                           "What Country or Region do you live in?")

df_2014 = get_column_data("/Users/rahulporuri/Downloads/SO-survey-data/2014_Stack_Overflow_Survey_Responses.csv",
                           "What Country do you live in?")

df_2016 = get_column_data("/Users/rahulporuri/Downloads/SO-survey-data/2016_Stack_Overflow_Survey_Responses.csv",
                           "country")

df_2017 = get_column_data("/Users/rahulporuri/Downloads/SO-survey-data/2017_Stack_Overflow_Survey_Responses.csv",
                           "Country")
```

```python
pd.concat([df_2011, df_2012, df_2013, df_2014, df_2016, df_2017], axis=1).dropna(axis=0, how="all")
```
