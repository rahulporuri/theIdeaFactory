```python
import glob
list_of_files = glob.glob("/Users/rahulporuri/Downloads/SO-survey-data/*")
```

```python
import pandas as pd
```

```python
pd.read_table("/Users/rahulporuri/Downloads/SO-survey-data/2011_Stack_Overflow_Survey_Results.csv",
               sep=",", header=0, engine="python", verbose=True)
```

```python
for file in list_of_files:
    print(file.split("/")[-1].split("_")[0], pd.read_table(file, sep=",", header=0, engine="python").shape[0])
```

```python
for file in list_of_files:
    print(file.split("/")[-1])
    df = pd.read_table(file, sep=",", header=0, engine="python")
    print(df.shape)
    try:
        col_wt_country_data = df.columns.str.lower().str.contains("country")
        country_data = df[df.columns[col_wt_country_data].get_values()[0]]
        print(country_data.astype("category").value_counts())  
    except:
        pass
```

```python
for file in list_of_files:
    df = pd.read_table(file, sep=",", header=0, engine="python")
    print(file.split("/")[-1], df.shape)
    col_wt_salary_data = df.columns.str.lower().str.contains("salary") | df.columns.str.lower().str.contains("compensation")
    try:
        print(df.columns[col_wt_salary_data].get_values())
        salary_data = df[df.columns[col_wt_salary_data].get_values()[0]]
        try:
            print(salary_data.astype("float").value_counts())
        except:
            print(salary_data.astype("category").value_counts())
    except:
        pass
```
